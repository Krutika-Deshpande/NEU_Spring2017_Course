```python
def email_analyse(inputfile, to_email_list, from_email_list, email_subj,email_body,message_ID):
    with open(inputfile, "r") as f:
        data = f.read()
    
    email = Parser().parsestr(data)
    
    from_email_list.append(email['from'])
    message_ID.append(email['Message-ID'])
    print("Message ID" + email['Message-ID'])
    #print(email['Message-ID'])
    email_subj.append(email['Subject'])
    #print(email_subj)
    email_body.append(email.get_payload())
    to_email_list.append(email['to'])
    row = str([cell.encode('utf-8') for cell in email_body])
    analysis=TextBlob(row)
   
    #print (analysis.sentiment) 
    if analysis.sentiment.polarity < 0:
        sentiment = "negative"
        
    
    elif analysis.sentiment.polarity == 0:
        sentiment = "neutral"
        
    else:
        sentiment = "positive"
        
    #print ("The sentiment of mails by Arnold John is:" ,+ str(sentiment))
        print(sentiment)
        
```

```python

with open("from_email_list.txt", "w") as f: # creating from emails text file
    for from_email in from_email_list:
        if from_email:
            f.write(from_email)
            f.write("\n")  
            
with open("email_subject.txt", "w") as f: # creating  email subject text file
    for email_sub in email_subj:
        if email_sub:
            f.write(email_sub)
            f.write("\n")
            

with open("email_body.txt", "w") as f:  # creating email body text file
    for email_bod in email_body:
        if email_bod:
            f.write(email_bod)
            f.write("\n")


with open("to_email_list.txt", "w") as f: # creating  to emails text file
    for to_email in to_email_list:
        if to_email:
            f.write(to_email)
            f.write("\n")
            
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords


with open("email_body.txt", "r") as f:
    data = f.read()
words = word_tokenize(data)
words2 = re.findall(r'(\b[a-z]{6,12}\b)',data )
useful_words = [word  for word in words2 if word not in stopwords.words('English')]
frequency = nltk.FreqDist(useful_words)
print("The most frequent words used in the email body:" )
print(frequency.most_common(10))
```
