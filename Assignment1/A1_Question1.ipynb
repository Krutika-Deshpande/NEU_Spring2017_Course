```python
import nltk
```

```python
nltk.corpus.gutenberg.fileids() #files in the Gutenberg package
```

```python
NumberOfWords = nltk.corpus.gutenberg.words('austen-persuasion.txt') # number of words in austen.persuasion.txt
len(NumberOfWords)

```

```python
print(NumberOfWords)
NumberOfUniqueWords = set(NumberOfWords) # set function is used to find number of unique words
print('The number of unique words in the austen-persuasion.txt' +  'are' + ' ' + str(len(NumberOfUniqueWords)))


```

```python
#from collections import Counter
from nltk.corpus import gutenberg
words = nltk.corpus.gutenberg.words('austen-persuasion.txt')
#freqs = Counter(words)   # using counter method from collections to find the frequency of words in the given text
#print('The frequency of each word in the austen-persuasion.txt is  ' +  str(freqs)) 

D = dict()
for a in words:
    if a not in D :
        D[a] = 1
    else:
        D[a] += 1
print(D)


```

```python
popular_words = sorted(D.items(), key = lambda x : x[1] ,reverse = True)
top_10 = popular_words[:10]
print('The top 10 most frequent words and their frequency is : ' + str(top_10))
```
